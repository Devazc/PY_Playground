```python
#Importing the requirerd packages 

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Read in the Data
df = pd.read_csv('D:\Code Here\Python Projects\Movie Recomendation System\Movie Correlation In Python\movies.csv')

# Let's take a look at the Data
df
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

**CORRELATION**

```python
# Correlation Matrix between all numeric columns
# PEARSON Correlation Coefficient.

df.corr(method = 'pearson', numeric_only=True)
```

```python
# KENDALL Correlation Coefficient.
# In Python, the scipy library provides functions to calculate Kendall's Tau.
# You can import the kendalltau function from the scipy.stats module to perform this calculation. 

from scipy.stats import kendalltau
df.corr(method='kendall', numeric_only=True)
```

```python
# SPEARMAN Correlation Coefficient

df.corr(method='spearman', numeric_only=True)
```

**CORRELATION MATRIX**

```python
correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Feature")

plt.xlabel("Movie Features")

plt.ylabel("Movie Features")

plt.show()
```

```python
# Using Factorize - this assigns a random numeric value for each unique category

df.apply(lambda x : x.factorize()[0]).corr(method = 'pearson', numeric_only=True)
```

```python
# CORRELATION MATRIX

correlation_matrix = df.apply(lambda x : x.factorize()[0]).corr(method='pearson', numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Movies")

plt.xlabel("Movie Features")

plt.ylabel("Movie Features")

plt.show()
```

**Analysis**

```python
correlation_matrix = df.apply(lambda x : x.factorize()[0]).corr()

corr_pairs = correlation_matrix.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind = "quicksort")

print(sorted_pairs)
```

```python
# The Pairs with High Correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
